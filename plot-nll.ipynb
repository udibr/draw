```python
import cPickle as pickle
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_df(name):
    with open(name + '/bmnist_log', "rb") as f:
        p = pickle.load(f)
    df = pd.DataFrame.from_dict(p, orient='index')
    df.index /= 500
    return df
```

python train-draw.py --attention=2,5 --niter=64 --lr=1e-3 --epochs=200 --bn

```python
df = load_df('bmnist-20150924-073426')
```

python train-draw.py --attention=2,5 --niter=64 --lr=3e-4 --epochs=100

```python
df0 = load_df('bmnist-20150922-180943')
```

```python
label='best=%.2f/%d'%(df.test_nll_bound.min(),df.test_nll_bound.argmin())
df.test_nll_bound.astype(float).iloc[1:].plot(label='lr=1e-3+BN '+label)
label0='best=%.2f/%d'%(df0.test_nll_bound.min(),df0.test_nll_bound.argmin())
df0.test_nll_bound.astype(float).iloc[1:].plot(label='lr=3e-4 '+label0)
plt.xlabel('epochs')
plt.ylabel('validation NLL')
plt.title('DRAW on binarized MNIST')
plt.legend();
```
